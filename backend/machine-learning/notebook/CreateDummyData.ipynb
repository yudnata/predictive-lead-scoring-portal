# Generate Dummy Data (Name, Telephone, email)

```python
import pandas as pd
import re
from faker import Faker

fake = Faker("id_ID")

def generate_email_from_name(name):
    if isinstance(name, str):
        patterns = (
            r'\b(?:Dr\.|Drs\.|Dra\.|Ir\.|Hj\.|H\.|Prof\.|'
            r'S\.Pd\.|S\.Pd\.I\.|S\.Kom\.|S\.T\.|S\.E\.|S\.Psi\.|S\.H\.|'
            r'M\.Pd\.|M\.M\.|M\.Hum\.|M\.Psi\.|M\.Ked\.|M\.Ak\.|M\.TI\.|'
            r'Dipl\.|A\.Md\.|A\.P\.|S\.Fil\.I\.|S\.Si\.|S\.Sos\.|S\.AB\.|'
            r'S\.I\.Kom\.|S\.Ag\.|S\.Pt\.|Dt\.|Haji|Hajjah|PhD|Ph\.D\.)\b'
        )

        name_clean = re.sub(patterns, '', name, flags=re.IGNORECASE)
        name_clean = re.sub(r'[\s,.]+', '.', name_clean).strip('.').lower()
        name_clean = re.sub(r'\.{2,}', '.', name_clean)

        if not name_clean:
            return fake.email()

        return f"{name_clean}@{fake.free_email_domain()}"

    return fake.email()

try:
    df = pd.read_csv('../data/bank-full.csv', delimiter=';')
except FileNotFoundError:
    print("File 'bank-full.csv' tidak ditemukan.")
    exit()

if "y" in df.columns:
    df = df.drop(columns=["y"])
    print("Kolom 'y' berhasil dihapus.")
else:
    print("Kolom 'y' tidak ditemukan, skip delete.")

num_rows = len(df)

unique_names = set()
unique_phones = set()
unique_emails = set()

names = []
phones = []
emails = []

def clean_phone_number(phone):
    if isinstance(phone, str):
        cleaned = re.sub(r'\D', '', phone)

        if cleaned.startswith('0'):
            cleaned = '62' + cleaned[1:]
        elif cleaned.startswith('62'):
            cleaned = cleaned
        elif len(cleaned) >= 8:
            cleaned = '62' + cleaned
        return cleaned
    return phone

def get_unique_name():
    while True:
        n = fake.name()
        if n not in unique_names:
            unique_names.add(n)
            return n

def get_unique_phone():
    while True:
        p = fake.phone_number()
        clean = clean_phone_number(p)
        if clean not in unique_phones:
            unique_phones.add(clean)
            return clean

def get_unique_email(name):
    while True:
        e = generate_email_from_name(name)
        if e not in unique_emails:
            unique_emails.add(e)
            return e

for _ in range(num_rows):
    name = get_unique_name()
    phone = get_unique_phone()
    email = get_unique_email(name)

    names.append(name)
    phones.append(phone)
    emails.append(email)

df['name'] = names
df['telephone'] = phones
df['email'] = emails

print("\nContoh data unik:")
print(df[['name', 'telephone', 'email']].head())

df.to_csv("../data/data_nasabah.csv", index=False)
print("\n✅ File berhasil disimpan sebagai 'data_nasabah.csv'")
```
